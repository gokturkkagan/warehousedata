```python
def get_one_hot_encoding(dataframe, categorical_columns):
    enc_oh = OneHotEncoder(sparse=False)
    enc_oh.fit(dataframe[categorical_columns])
    category_columns = np.concatenate(enc_oh.categories_)
    encoded_features = enc_oh.transform(dataframe[categorical_columns])
    dataframe[category_columns] = pd.DataFrame(
        encoded_features,
        columns=category_columns,
        index=dataframe.index)
    return dataframe.drop(categorical_columns, axis=1)
```

```python
def visualize_regression_results(Y_test, Y_predicted, axes, model=None):
    """
    Plots two prediction - ground truth comparisons for model evaluation.
    
    The first plot compares the predicted values against the real values
    for the list of test labels. The ground truth - prediction pairs are
    added as rows to a dataframe. The dataframe is then sorted by the 
    ground truth values, melted using pd.melt and plottet using a seaborn
    lineplot.
    
    The second grafic plots the real target variable values (x-axis) 
    against the generated predictions using a seaborn scatterplot. Then 
    the function x=y is added to the plot
    
    :param Y_test: The list of ground truth labels.
    :param Y_predicted: The list of corresponding predictions.
    :param axes: A list of two matplotlib axes to plot to.
    :param model: A string denoting the model used to generate predictions.
    
    :return: The modified axes
    """
    df_prediction = pd.DataFrame({
        'Actual':Y_test, 
        'Predicted':Y_predicted
    }).sort_values('Actual').reset_index(drop=True)
    df_prediction['Test Set Example'] = df_prediction.index

    df_lineplot = pd.melt(df_prediction, id_vars=['Test Set Example'],
                          value_vars=['Actual', 'Predicted'],
                          var_name='Curve', value_name='Target Variable')
    ax_lineplot = sns.lineplot(data=df_lineplot.sort_index(ascending=False, axis=0), 
                               y='Target Variable', x='Test Set Example', 
                               hue='Curve', style='Curve', ax=axes[0])
    ax_lineplot.set_title(model + ' Results:\n'
                          'Predictions for Test Set Examples \n'
                          'Sorted by the Corresponding Groud Truth Values', loc='left')
    
    xlims=(0,15)
    ax_scatter = sns.scatterplot(x=Y_test, y=Y_predicted, ax=axes[1], color='r')
    ax_scatter.set_xlabel('Actual')
    ax_scatter.set_ylabel('Predicted')
    ax_scatter.set_title(' \n' # \textbf is latex for bold ;)
                         'Predictions -- Ground Truth Scatterplot\n'
                         'Ideal Situation (Predicted = Actual) in Green', loc='left')
    
    ax_scatter.plot(xlims, xlims, color='g')
    return ax_lineplot, ax_scatter
```

```python
def plot_learning_curves(model, x, y, cv=3, scoring='accuracy',
                         scoring_unit='', model_name='',
                         fmt='png', title='',
                         train_sizes=np.linspace(0.1, 1.0, 10),
                         n_jobs=1, ax=None):

    # get scores
    try:
        ds_sizes, train_scores, valid_scores = load(
            open('./data/lc_rgr_' + model_name + '.pickle', 'rb'))
    except FileNotFoundError:
        ds_sizes, train_scores, valid_scores = learning_curve(
            model, x, y, scoring=scoring, n_jobs=n_jobs,
            train_sizes=train_sizes, cv=cv, shuffle=True)
        dump((train_sizes, train_scores, valid_scores), 
             open('./data/lc_rgr_' + model_name + '.pickle', 'wb'))
    
    # create dataframe for plot
    mean_train_scores = np.mean(train_scores, axis=1)
    mean_valid_scores = np.mean(valid_scores, axis=1)
    score = prettify_scoring_name(scoring)
    scores_df = pd.DataFrame({
        'Training Set Size': np.concatenate((ds_sizes, ds_sizes)),
        score: np.concatenate(
            (mean_train_scores, mean_valid_scores)),
        'Dataset': (['Training'] * train_sizes.shape[0] + 
                    ['Validation'] * train_sizes.shape[0])
    })

    # create plot
    if not title:
        title = (model_name + '\n'
                 'Average Training and Validation Errors Relative to Training ' 
                 'Set Sizes\n(' + str(cv) + '-Fold Cross Validation Was Used)')
    axes = sns.lineplot(
        x='Training Set Size', y=score, hue='Dataset', data=scores_df, ax=ax)
    axes.set_title(title, loc='left')
    

def prettify_scoring_name(scoring_name):
    """
    Takes a scorer name, splits it into component words (on "_"), capitalizes
    those and returns the string joined by " ".

    :param scoring_name: The scorer name with "_" as word separators.
    :return: The scorer name with " " as (capitalized) word separators.
    """
    words = scoring_name.split('_')
    return " ".join(list(map(lambda x: x.capitalize(), words)))

```
